```python
#importing dependancies
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
#loading data straighr from url
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

```

```python
print(data)
```

```python
# Convert 'data' array to DataFrame
data_df = pd.DataFrame(data)

# Convert 'target' array to DataFrame with appropriate column name
target_df = pd.DataFrame(target, columns=['Target'])
```

```python
#getting some statistics from the datasets
data_df.describe()
```

```python
#initialize the linear regression model
reg = linear_model.LinearRegression()
```

```python
# split the data into 80% training and 20% testing data
X_train, X_test, y_train, y_test = train_test_split(data_df, target_df, test_size= 0.2, random_state =42)
```

```python
# train the model with our training data
reg.fit(X_train, y_train)
```

```python
#print the coefficient for each column of our model
print(reg.coef_)
```

```python
#print the prediction on our test data
y_pred = reg.predict(X_test)
print(y_pred)
```

```python
#print actual values
print(y_test)
```

```python
# check the model perfomance/accuracy using MSE
print(np.mean( (y_pred - y_test)**2))
```

```python
# check the model perfomance/accuracy using MSE and sklearn_metrics
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred))
```
